## save data per plane

```python
in_file='../data/'
out_file='../data_per_plane/'
listfile=os.listdir('../data')
plane=['P123','P124','P125','P126','P127']
for p in plane:
    DATA,M,S,number=pd.DataFrame(),pd.DataFrame(),pd.DataFrame(),0
    for i in listfile:
        if i.startswith(p):
            data = pd.read_table(in_file+i, sep=' ', encoding='gbk',header=None)
            data = data.drop(labels=36, axis=1)
            # mean
            m = pd.DataFrame(np.array((data.loc[0]+1000).tolist()).reshape(1,36))
            m.insert(0,'number',[number]*m.shape[0])
            M = pd.concat([M,m],ignore_index=True) 
            # standard deviation
            s = pd.DataFrame(np.array((data.loc[1]+1000).tolist()).reshape(1,36))
            s.insert(0,'number',[number]*s.shape[0])
            S = pd.concat([S,s],ignore_index=True) 
            # feature data
            data = data.drop(index=[0,1]).reset_index(drop=True)
            data.insert(0,'number',[number]*data.shape[0])
            DATA = pd.concat([DATA,data],ignore_index=True) 
            number=number+1
    DATA.to_csv(out_file+p+'_data.csv',index=False,sep=',')
    M.to_csv(out_file+p+'_mean.csv',index=False,sep=',')
    S.to_csv(out_file+p+'_sd.csv',index=False,sep=',')
```
