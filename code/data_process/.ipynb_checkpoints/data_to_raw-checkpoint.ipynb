```python
import pandas as pd
```

```python
columnName = ["时间","全机油量","武器重量","马赫数","气压高度","校准空速","真空速","升降速度","攻角","侧滑角","动压","法向过载","侧向过载","轴向过载","俯仰角","滚转角","航向角","滚转速率","俯仰速率","偏航速率","左鸭翼","右鸭翼","左前襟","右前襟","左外副翼","右外副翼","左内副翼","右内副翼","左方向舵","右方向舵","机翼剪力","机翼弯矩","鸭翼剪力","垂尾剪力","鸭翼弯矩","机身弯矩"]
```

```python
title = "1.dat"
```

```python
file = pd.read_csv(title,delim_whitespace=True,header=None)
```

```python
file
```

```python
columnSize = len(file.columns)
```

```python
columnSize
```

```python
rowSize = len(file.iloc[:,0])
```

```python
rowSize
```

```python
tmpDic = {columnName[i]:[]for i in range(columnSize)}
```

```python
for k in range(columnSize):
    mean = file.iloc[0,k]+1000
    std = file.iloc[1,k]+1000
    for j in range(rowSize-2):
        tmpDic[columnName[k]].append(file.iloc[j+2,k]*std + mean)
```

```python
data = pd.DataFrame(tmpDic)
```

```python
data
```

```python
for k in range(columnSize):
    if(k==2):
        continue
    maxinum = max(data.iloc[2:,k])
    mininum = min(data.iloc[2:,k])
    #print(maxinum-mininum)
    print(columnName[k]+":"+str((maxinum-mininum)/(file.iloc[1,k]+1000)))
```

```python
import numpy as npy
import statsmodels.api as sm
```

```python
dat = data.iloc[:,12]
ecdf = sm.distributions.ECDF(dat)
#等差数列，用于绘制X轴数据
x = npy.linspace(min(dat), max(dat))
# x轴数据上值对应的累计密度概率
y = ecdf(x)
#绘制阶梯图
plt.step(x, y)
plt.show()
```
